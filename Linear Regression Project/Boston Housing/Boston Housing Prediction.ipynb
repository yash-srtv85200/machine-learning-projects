# Splitting the dataset 

```python
x = df.drop('Price', axis=1)
y = df['Price']
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Applying Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

# These are the values of B1, B2 --- in linear equation
for i, name in enumerate(x.columns):  
    print(f'{name}: {model.coef_[i]}')
```

```python
# Lets see the value of intercept
#print(f'Intercept: {model.intercept_}')
print(model.intercept_)
```

```python
#OLS --- Ordinary Least Squares
#In the result :
#No. Observations: Means the total number of observations (data points) used in the model fitting.
#Df Residuals: Degrees of freedom of the residuals, calculated as the number of observations minus the number of estimated parameters (including the intercept).
#Df Model: The number of estimated parameters in the model, excluding the intercept.
#R-squared: A statistical measure that represents the proportion of variance for the dependent variable that's explained by the independent variables in the model. Values range from 0 to 1, with higher values indicating a better fit.
#Adj. R-squared: Adjusted R-squared adjusts the R-squared value for the number of predictors in the model. It is a more accurate measure when comparing models with different numbers of predictors.
#F-statistic: A value that tests the overall significance of the model. It compares the model with no predictors to the specified model. A higher value indicates that the model is statistically significant.
#Prob (F-statistic): The p-value corresponding to the F-statistic. A small p-value (typically < 0.05) indicates that at least one of the predictors is statistically significant.
#Log-Likelihood: A measure of the likelihood of the model given the data. It is used in various statistical tests and model comparisons.
#AIC: A criterion for model selection among a finite set of models. Lower values indicate a better fit relative to the number of parameters in the model.
#BIC: Similar to AIC, but it applies a larger penalty for models with more parameters. Again, lower values indicate a better model fit.
#Omnibus: A test for normality of the residuals. A low p-value suggests that the residuals are not normally distributed.
#Durbin-Watson: A statistic that tests for the presence of autocorrelation in the residuals. Values close to 2 indicate no autocorrelation, while values below 1 or above 3 suggest potential issues.
#Jarque-Bera (JB): A test statistic for normality. It assesses whether the sample data has the skewness and kurtosis matching a normal distribution.
#Cond. No.: The condition number, which indicates potential multicollinearity problems. A high value (typically > 30) suggests that multicollinearity may be an issue.
```

```python
import statsmodels.api as sm

# Add constant term for the intercept
X_train_with_constant = sm.add_constant(X_train)  

# Fit the OLS model
ols_model = sm.OLS(y_train, X_train_with_constant).fit()

# Print the summary of the regression
print(ols_model.summary())
```

# Testing the accuracy of the model

```python
predicted = model.predict(X_test)
actual = y_test
```

```python
df_result = pd.DataFrame()
df_result['Actual'] = pd.Series(actual)
df_result['Predicted'] = pd.Series(predicted)
df_result = pd.DataFrame({'Actual':actual, 'Predicted': predicted})
df_result.head()
```

```python
df_result['Residuals'] = df_result['Actual'] - df_result['Predicted']
df_result
```

```python
# Linear Regression Assumption : If we try to plot the errors, the plotted chart should be normally distributed.
# Q-Q Plot is used to check if you data is in a normal distribution
plt.figure(figsize=(8, 6))
sns.histplot(df_result['Residuals'], bins=30, kde=True)
plt.xlabel('Residuals')
plt.title('Distribution of Residuals')
plt.grid()
plt.show()
```

```python
# Set the style for Seaborn
sns.set(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_result, x='Actual', y='Predicted', marker='o', color='blue', s=100, label='Predicted Values')

# Create a dashboard line (45-degree line)
min_val = df_result[['Actual', 'Predicted']].min().min()
max_val = df_result[['Actual', 'Predicted']].max().max()
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', 
         linewidth=5, label='Ideal Prediction Line')

# Enhance the plot with labels and titles
plt.xlabel('Actual Values', fontsize=14)
plt.ylabel('Predicted Values', fontsize=14)
plt.title('Scatter Plot of Actual vs Predicted Values', fontsize=16)
plt.legend()
plt.xlim(min_val - 1, max_val + 1)  
plt.ylim(min_val - 1, max_val + 1)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# R^2 is the measure of goodnessof fit. It explain "how well the variable is explained by the model"
# R^2 closer to 1 is considered as good.
from sklearn import metrics
print(f'The R2 score: {metrics.r2_score(actual,predicted)}\n')
print(f'The MSE: {metrics.mean_squared_error(actual,predicted)}')
```

```python
X = df.drop('Price', axis=1)  
y = df['Price']  
```

```python
from sklearn.linear_model import ElasticNet, Lasso, Ridge, LinearRegression
from sklearn.model_selection import KFold, cross_val_score

# Define estimators
estimators = {
    'LinearRegression': LinearRegression(),
    'ElasticNet': ElasticNet(),
    'Lasso': Lasso(),
    'Ridge': Ridge()
}

# Initialize KFold
kfold = KFold(n_splits=10, random_state=11, shuffle=True)

# Evaluate each estimator
for estimator_name, estimator_object in estimators.items():
    scores = cross_val_score(estimator=estimator_object, X=X, y=y, cv=kfold, scoring='r2')
    print(f'{estimator_name:>16}: mean of R² scores = {scores.mean():.3f}')
```
